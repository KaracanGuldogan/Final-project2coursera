```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# WebDriver'i başlat
driver = webdriver.Chrome()  # veya uygun tarayıcı sürücüsünü seçin
driver.get("https://finance.yahoo.com/quote/TSLA/financials")

# Sayfanın yüklenmesini bekle
time.sleep(5)

# Tabloyu bul
table = driver.find_element(By.XPATH, "//table")
html_data = table.get_attribute('outerHTML')

# Pandas ile tabloyu dataframe'e çevir
tesla_revenue = pd.read_html(html_data)[0]
print(tesla_revenue)

driver.quit()
```

```python
import yfinance as yf
import pandas as pd

# GameStop için Ticker objesi oluştur
gme = yf.Ticker("GME")

# Hisse senedi verilerini çek
gme_data = gme.history(period="max")

# İndeksi sıfırla
gme_data.reset_index(inplace=True)

# İlk 5 satırı görüntüle
print(gme_data.head())
```

### 4. Extracting GameStop Revenue Data Using Webscraping


```python
import requests
from bs4 import BeautifulSoup

# Web sayfasını indir
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text

# HTML'yi parse et
soup = BeautifulSoup(html_data, "html.parser")

# Tabloyu bul ve veriyi çıkar
gme_revenue = pd.read_html(str(soup.find_all("table")[1]))[0]

# Kolon adlarını kontrol et ve düzenle
gme_revenue.columns = ["Date", "Revenue"]

# Gelir sütunundan $, , karakterlerini kaldır
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(',', "").str.replace('$', "")

# Null ve boş değerleri temizle
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]

# Son 5 satırı görüntüle
print(gme_revenue.tail())
```
